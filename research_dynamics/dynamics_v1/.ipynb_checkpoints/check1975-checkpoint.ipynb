# helper function for comparing journals present in 2 sets of pmids

```python

from collections import defaultdict


def compare_journal_difference(pmids1, pmids2):
    '''
    return score for distance between pmids in terms of journal similarity
    note that there are name resolution issues for the journal names
      so sets will be more dissimilar than in a perfectly resolved db
    
    '''
    curs1 = get_journals_for_pmids(pmids1)
    curs2 = get_journals_for_pmids(pmids2)
    
    d1 = get_token_counts(curs1.fetchall())
    d2 = get_token_counts(curs2.fetchall())
    
    score = get_journal_distance_score(d1, d2)
    print("difference score: ".format(score))
    print_biggest_differences(d1, d2)
    
    return score
    
    
def get_journals_for_pmids(pmid_set):
    '''
    read in a list of pmids (not a set)
    connect to the database (configured above)
    return cursor over list of journals corresponding to pmids
    '''
    str_fmt = ', '.join([str(pmid) for pmid in pmid_set])

    sql = '''SELECT M.journal
            FROM metadata as M
            WHERE M.pmid IN ({})'''.format(str_fmt)

    cursor = db.cursor()
    cursor.execute(sql)
    
    return cursor


def get_token_counts(token_list):
    '''
    number of occurrences for each unique token in the input list
    d_count output is a dictionary such that d_count[token] = #occurrences
    '''
    d_count = defaultdict(int)
    for token in token_list:
        d_count[token] += 1
    return d_count

def get_journal_distance_score(d1, d2):
    '''
    d1,d2 - dictionary of token counts 1 and 2
    return L1 distance for counts over each unique key
    '''
    journals1 = [k for k in d1.keys()]
    journals2 = [k for k in d2.keys()]
    all_journals = set(journals1).union(set(journals2))
    score = 0
    for k in all_journals:
        L1_distance = abs(d1[k] - d2[k])
        score += L1_distance
    return score

def print_biggest_differences(d1, d2, TOP=10):
    '''
    this is redundant with get_journal_distance_score
       so if speed becomes a problem take care of that
       print out the keys and scores for the most dissimilar and similar
           journals/counts
    '''
    journals1 = [k for k in d1.keys()]
    journals2 = [k for k in d2.keys()]
    all_journals = set(journals1).union(set(journals2))
    D = defaultdict(int)
    for k in all_journals:
        L1 = abs(d1[k] - d2[k])
        D[k] = L1
    
    print()
    # find largest 
    k_largest = sorted(D, key=D.get, reverse=True)[:TOP]
    print('-----least similar-----')
    print('journal,               count1, count2')
    for k in k_largest:
        print(k, d1[k], d2[k])
    print()
    # find smallest
    print('-----most similar-----')
    print('journal,               count1, count2')
    k_smallest = sorted(D, key=D.get)[:TOP]
    for k in k_smallest:
        print(k, d1[k], d2[k])
    print()
    
```
