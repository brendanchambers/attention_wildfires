# write evaluation data to file for simpler IO during batching

```python
sql = 'SELECT pmid, title, abstract FROM abstracts'

db = mysql.connect(**client_config)
cursor = db.cursor()
start_time = time.time()
cursor.execute(sql)
print('elapsed time: {}'.format(time.time() - start_time))

start_time = time.time()

abstracts_per_file = 600000
data_chunk = []
chunk_id = 0
for idx, row in enumerate(cursor):
    
    data_chunk.append(row)
    
    if idx % abstracts_per_file == 0:  # note - first file with only have 1 abstract
        print('writing chunk to file...')
        # write data chunk to new file
        
        with open(target_path + 'datachunk_{}.csv'.format(chunk_id),'w') as f:
            csv_out = csv.writer(f, delimiter=' ')
            csv_out.writerow(['pmid','title','abstract'])
            for row_ in data_chunk:
                csv_out.writerow(row_)
                
        # reset data_chunk
        print('finished chunk {}'.format(chunk_id))
        chunk_id += 1
        data_chunk = []
        print('parsing from cursor...')
    
print('elapsed time: {}'.format(time.time() - start_time))

cursor.close()
db.close()
```
