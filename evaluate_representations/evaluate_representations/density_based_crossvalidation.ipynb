```python
# helper function for cross-validation


def partition_samples(d_embeddings):

    d_partition = {}
    N_partitions = 3
    N_journals = 2  # todo need to change this to support more journals
    
    for journal in journals:
        d_partition[journal] = []
        this_journal = d_embeddings[journal].copy()

        L_partition = int(math.floor((1.0/N_partitions)*len(this_journal)))
        #print( (len(this_journal), L_partition))
        random.shuffle(this_journal)  # shuffle in place

        curPosition = 0
        for partition in range(N_partitions):
            partition_samples = this_journal[curPosition:(curPosition+L_partition)]            
            curPosition += L_partition
            #print(curPosition)
            d_partition[journal].append(partition_samples)
            
            doPlotPartition=False  # switch for sanity check plotting
            if doPlotPartition:
                plt.figure(figsize=(2,2))
                plt.scatter(d_partition[journal][partition][:,0],
                           d_partition[journal][partition][:,1],
                           marker='.',
                           alpha=0.25)
                plt.title(partition)
                
    return d_partition

def emd_xvalidation(d_partition):
    
    N_journals = len(d_partition.keys())
    N_partitions = len(d_partition[journals[0]])
    
    # within
    within_scores = {}
    for i_journal in range(N_journals):
        journal = journals[i_journal]
        within_scores[journal] = []
        
    for i_partition in range(N_partitions):
        for j_partition in range(i_partition+1,N_partitions):
            
            for i_journal in range(N_journals):
                journal = journals[i_journal]
                P = d_partition[journal][i_partition]
                Q = d_partition[journal][j_partition]
                score = emd_samples(P,Q)
                within_scores[journal].append(score)
        
    # between
    between_scores = []
    journal_i = journals[0]
    journal_j = journals[1]
    
    for i_partition in range(N_partitions):
        for j_parititon in range(N_partitions):
            P = d_partition[journal_i][i_partition]
            Q = d_partition[journal_j][j_partition]
            score = emd_samples(P,Q)
            between_scores.append(score)
                        
    return (within_scores, between_scores)
         

def cos_sim_helper(d_embeddings, title_text='cosine similarities by class'):
    
    ## partition samples in embedding dictionary, for cross-validation 
    #condition = embedding_conditions[0]
    d_partition = {}
    N_partitions = 2  # currently not supporting other choices of N_partitions

    for journal in journals:
        d_partition[journal] = []
        this_journal = d_embeddings[journal].copy()

        L_partition = int(math.ceil((1.0/N_partitions)*len(this_journal)))
        random.shuffle(this_journal)  # shuffle in place

        curPosition = 0
        for partition in range(N_partitions):
            partition_samples = this_journal[curPosition:(curPosition+L_partition)]
            curPosition = L_partition
            d_partition[journal].append(partition_samples)
        

    ### measure cosine similarity within and between samples
    cos_sim_xvalid = {}

    my_colors = [[[0,0.5, 0.5],[0.75, 0.25, 0.5]],
                 [[0.75, 0.5, 0.5], [0, 0.75, 0.25]]]

    plt.figure()

    for i_src, src_journal in enumerate(journals):
        src_samples = d_partition[src_journal][0]

        for i_tgt, tgt_journal in enumerate(journals):
            tgt_samples = d_partition[tgt_journal][1]

            R = cosine_similarity(src_samples, tgt_samples)
            cos_sim_xvalid[(src_journal, tgt_journal)] = R


            plt.hist(R.flatten(),
                     density=True,
                     histtype='step',
                     bins=100,
                     color=my_colors[i_src][i_tgt],
                     alpha=0.5,
                     linewidth=2)
            
            plt.xlabel('cosine similarity')
            plt.ylabel('density')
            plt.title(title_text)
            
    plt.savefig('pca xvalidation 2D {}.png'.format(title_text))
    plt.savefig('pca xvalidation 2D {}.svg'.format(title_text))
            
    return cos_sim_xvalid

```

```python
# test resampling from the kde

def resample(kde, size):  # source: https://stackoverflow.com/questions/50520535/resampling-a-kde-kernel-density-estimation-in-statsmodels
    n, d = kde.data.shape
    indices = np.random.randint(0, n, size)
    cov = np.diag(kde.bw)**2
    means = kde.data[indices, :]
    norm = np.random.multivariate_normal(np.zeros(d), cov, size)
    return np.transpose(means + norm)
```

```python
# test generating a meshgrid

def gen_grid(kde):
    n, d = kde.data.shape
    
    min_vals = np.min(kde.data,0)
    max_vals = np.max(kde.data,0)
    
    params = []
    for i in range(d):
        xx = np.arange(start=min_vals[i], stop=max_vals[i], step=kde.bw[i])
        print(np.shape(xx))
        params.append(xx)
        
    grid = np.meshgrid(*params, indexing='xy')
    return grid
```
